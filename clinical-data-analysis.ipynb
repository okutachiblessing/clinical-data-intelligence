## Section 5: Prepare Data for Visualization

Format and aggregate data into structures suitable for Power BI dashboards

```python
# Data Quality Metrics
print("=" * 80)
print("DATA QUALITY ASSESSMENT")
print("=" * 80)

# 1. Duplicate Detection
duplicates_patients = df_patients.duplicated(subset=['patient_id']).sum()
duplicates_measurements = df_measurements.duplicated(subset=['visit_id']).sum()

print(f"\n[DUPLICATES]")
print(f"  Patient duplicates: {duplicates_patients}")
print(f"  Measurement duplicates: {duplicates_measurements}")

# Remove duplicates
df_patients = df_patients.drop_duplicates(subset=['patient_id'], keep='first')
df_measurements = df_measurements.drop_duplicates(subset=['visit_id'], keep='first')

# 2. Missing Values Analysis
print(f"\n[MISSING VALUES]")
missing_patients = df_patients.isnull().sum()
print("\nPatient Data Missing Values:")
print(missing_patients[missing_patients > 0])

missing_measurements = df_measurements.isnull().sum()
print("\nMeasurements Data Missing Values:")
print(missing_measurements[missing_measurements > 0])

# 3. Data Type Conversion and Validation
print(f"\n[DATA TYPE CONVERSION]")

# Convert gender to standardized format
df_patients['gender'] = df_patients['gender'].str.upper().map({
    'M': 'Male', 'F': 'Female', 'O': 'Other'
})

# Remove whitespace from phone numbers, validate format
if 'phone_number' in df_patients.columns:
    df_patients['phone_number'] = df_patients['phone_number'].str.replace(r'[^\d]', '', regex=True)
    df_patients['phone_number'] = df_patients['phone_number'].apply(
        lambda x: x if len(x) == 10 else np.nan
    )

# Calculate patient age
df_patients['age'] = (pd.to_datetime('today') - df_patients['date_of_birth']).dt.days // 365

# 4. Clinical Data Validation
print(f"\n[CLINICAL DATA VALIDATION]")

# Define valid ranges for vital signs
clinical_ranges = {
    'systolic_bp': (70, 220),
    'diastolic_bp': (40, 140),
    'heart_rate': (30, 200),
    'temperature': (95.0, 106.0),
    'glucose': (40, 500),
}

# Flag outliers and replace with median
outlier_counts = {}
for measurement, (min_val, max_val) in clinical_ranges.items():
    if measurement in df_measurements.columns:
        outliers = ((df_measurements[measurement] < min_val) | 
                   (df_measurements[measurement] > max_val)).sum()
        outlier_counts[measurement] = outliers
        
        # Replace outliers with NaN then impute with median
        df_measurements.loc[
            (df_measurements[measurement] < min_val) | 
            (df_measurements[measurement] > max_val), 
            measurement
        ] = np.nan
        
        median_val = df_measurements[measurement].median()
        df_measurements[measurement].fillna(median_val, inplace=True)

print("Outliers detected and corrected:")
for measurement, count in outlier_counts.items():
    print(f"  {measurement}: {count} outliers")

# 5. Create derived features
print(f"\n[FEATURE ENGINEERING]")

# Mean Arterial Pressure (MAP)
df_measurements['mean_arterial_pressure'] = (
    df_measurements['systolic_bp'] + 
    2 * df_measurements['diastolic_bp']
) / 3

# Pulse Pressure
df_measurements['pulse_pressure'] = (
    df_measurements['systolic_bp'] - 
    df_measurements['diastolic_bp']
)

# Risk Tier Classification
def classify_risk_tier(row):
    risk_score = 0
    if pd.notna(row.get('systolic_bp', 0)) and row['systolic_bp'] > 180:
        risk_score += 2
    if pd.notna(row.get('glucose', 0)) and row['glucose'] > 200:
        risk_score += 2
    if pd.notna(row.get('heart_rate', 0)) and (row['heart_rate'] > 100 or row['heart_rate'] < 50):
        risk_score += 1
    
    if risk_score >= 4:
        return 'High'
    elif risk_score >= 2:
        return 'Medium'
    else:
        return 'Low'

df_measurements['risk_tier'] = df_measurements.apply(classify_risk_tier, axis=1)

print(f"  ✓ Mean Arterial Pressure calculated")
print(f"  ✓ Pulse Pressure calculated")
print(f"  ✓ Risk tier classification complete")

# 6. Date Features for Time-Series Analysis
df_measurements['visit_year'] = df_measurements['visit_date'].dt.year
df_measurements['visit_month'] = df_measurements['visit_date'].dt.month
df_measurements['visit_day_of_week'] = df_measurements['visit_date'].dt.day_name()
df_measurements['visit_week'] = df_measurements['visit_date'].dt.isocalendar().week

print(f"  ✓ Time-series features extracted")

print(f"\n✓ Data processing complete!")
print(f"  Patient records: {len(df_patients)}")
print(f"  Clinical measurements: {len(df_measurements)}")
```
